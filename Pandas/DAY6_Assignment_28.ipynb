```python
import pandas as pd
```

```python
# 1. Load the csv dataset “Netflix TV Shows and Movies” into a DataFrame calledratings.
ratings = pd.read_csv("./Netflix.csv")
ratings
```

```python
# 2. Select the description column from ratings and assign the result to the variable
# desc. What type of object is desc?

desc = ratings["description"]
desc
type(desc) #it is a object of type Series
```

```python
# 3. Select the first value from the description column of ratings, assigning it to variable first_description.

first_description = ratings["description"].loc[0]
print(first_description)
```

```python
# 4. Select the first record from ratings, assigning it to the variable first_row.

first_row = ratings.iloc[[0]]
first_row
```

```python
# 5. Select the first 15 values from the description column in ratings, assigning the result to variable first_descriptions.
first_descriptions = ratings["description"].head(15)
print(first_descriptions)
```

```python
# 6. Select the records with index labels 1, 2, 4, 7, and 11, assigning the result to the variable sample_ratings.
sample_ratings = ratings.loc[[1, 2, 4, 7, 11]]
sample_ratings
```

```python
# 7. Create a DataFrame df containing the title, type, release_year, and age_certification, columns of 
# the records with the index labels 0, 4, 50, and 200.

df = ratings.loc[[0, 4, 50, 200],["title", "type", "release_year", "age_certification"]]
df
```

```python
# 8. Create a DataFrame df containing the title and release_year columns of the first 50 records.
df = ratings[["title","release_year"]].head(50)
df
```

```python
# 9. Create a DataFrame show_ratings containing ratings of shows.

show_ratings = ratings[["title","imdb_score"]]
show_ratings
```

```python
# 10. Create a DataFrame great_shows containing shows with ratings more than 7.5.

mask = ratings["imdb_score"] > 7.5
great_shows = ratings.loc[mask]
great_shows
```

```python
# 11. Create a DataFrame mini containing movies with runtime less than an hour.

mask = ratings["runtime"] < 100
mini = ratings.loc[mask]
mini
```

```python
# 12. Create a DataFrame my_age containing movies that were released in your year of birth.

mask = (ratings["release_year"] == 2003)
my_age = ratings.loc[mask]
my_age
```

```python
# 13. Create a DataFrame my_movies containing movies that were released in the year 2000,
# 2004, 2008 and 2012.

mask = (ratings["release_year"] == 2000) | (ratings["release_year"] == 2004) | (ratings["release_year"] == 2008) | (ratings["release_year"] == 2012)
my_movies = ratings.loc[mask]
my_movies
```

```python
# 14. What is the mean, median and mode of the imdb_score column in the ratings DataFrame?
print(f"Mean of the imdb_score is : {ratings['imdb_score'].mean()}")
print(f"Median of the imdb_score is : {ratings['imdb_score'].median()}")
print(f"Mode of the imdb_score is : {ratings['imdb_score'].mode()}")
```

```python
# 15. What release years are represented in the dataset (must not include duplicates)?

ratings["release_year"].unique()
```

```python
# 16. How often does each release year appear in the dataset?

ratings["release_year"].value_counts()
```

```python
# 17. What are the most common runtime of the movies in the dataset?

mcrt = ratings["runtime"].value_counts().sort_values(ascending=False)
mcrt

```

```python
# 18. Which are the top three mostly voted movies?

ratings.nlargest(3,"imdb_votes")
```
